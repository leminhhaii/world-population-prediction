```python
import pandas as pd
```

```python
gdp_per_capita = pd.read_csv("GDP_per_Capita.csv")
life_expectancy = pd.read_csv("Life Expectancy.csv")
male_ratio = pd.read_csv("Male Population Ratio.csv")
population = pd.read_csv("Population.csv")
old_ratio = pd.read_csv("Retirement Age Dependency Ratio.csv")
area = pd.read_csv("Total Area (sq km).csv")
urban_population_ratio = pd.read_csv("Urban Population Ratio.csv")
sdg_regions = pd.read_csv("world_regions_sdg.csv")
young_ratio = pd.read_csv("Young Age Dependency Ratio.csv")
```

```python
infant_mortality = pd.read_csv("Infant Mortality.csv")
infant_mortality.drop(columns="Variant",inplace=True)
infant_mortality.rename(columns={"Value":"Infant mortality per 1000 live births"},inplace=True)
```

```python
country_codes = pd.read_csv("country_codes.csv")
country_codes = country_codes.drop(columns= "Code")
country_codes['ISO3'] = country_codes['ISO3'].replace('ROM', 'ROU')
country_codes['ISO3'] = country_codes['ISO3'].replace('ZAR', 'COD')
country_codes['ISO3'] = country_codes['ISO3'].replace('TMP', 'TLS')
new_data = {
    'ISO3': ['MNE', 'SRB'],
    'Country': ['Montenegro', 'Serbia']
}
new_countries = pd.DataFrame(new_data)
for _, row in new_countries.iterrows():
    iso3 = row['ISO3']
    country = row['Country']
    if iso3 not in country_codes['ISO3'].values and country not in country_codes['Country'].values:
        country_codes = pd.concat([country_codes, pd.DataFrame([row])], ignore_index=True)
country_codes
```

```python
rawbmi = pd.read_csv("bmi_data.csv")
rawbmi = rawbmi[rawbmi["SpatialDimType"] == 'COUNTRY']
rawbmi = rawbmi[["SpatialDim","Dim1","TimeDim","NumericValue"]]
rawbmi = rawbmi.pivot_table(
    index=['TimeDim', 'SpatialDim'], 
    columns='Dim1', 
    values='NumericValue'
).reset_index()
rawbmi.rename(columns={'TimeDim': 'Year', 'SpatialDim': 'ISO3','SEX_BTSX':"BMI","SEX_FMLE":"BMI Female","SEX_MLE":"BMI Male"}, inplace=True)
rawbmi['ISO3'] = rawbmi['ISO3'].replace('SDN736', 'SDN')
rawbmi
```

```python
bmi = pd.merge(country_codes,rawbmi,on="ISO3",how='inner')
bmi = bmi.drop(columns="ISO3")
bmi
```

```python
def changeCountryName(oldnames, new_name):
    datasets = [
        (bmi, 'Country'),
        (gdp_per_capita, 'Country'),
        (life_expectancy, 'Country'),
        (male_ratio, 'Country'),
        (old_ratio, 'Country'),
        (young_ratio, 'Country'),
        (population, 'Country'),
        (urban_population_ratio, 'Country'),
        (area, 'Country'),
        (infant_mortality, 'Country or Area'),
        (sdg_regions, 'Country/area')
    ]
    for df, column in datasets:
        if column in df.columns:
            df[column].replace(oldnames, new_name, inplace=True)
```

```python
country_bmi = set(bmi.Country.unique())
country_gdp = set(gdp_per_capita.Country.unique())
country_life = set(life_expectancy.Country.unique())
country_male = set(male_ratio.Country.unique())
country_old = set(old_ratio.Country.unique())
country_young = set(young_ratio.Country.unique())
country_population = set(population.Country.unique())
country_urban = set(urban_population_ratio.Country.unique())
country_area = set(area.Country.unique())
country_infant = set(infant_mortality["Country or Area"].unique())
country_sdg = set(sdg_regions['Country/area'].unique())
```

```python
common_countries = country_bmi.intersection(country_bmi,country_gdp, country_life, country_male, 
                                             country_old, country_young, 
                                             country_population, country_urban, 
                                             country_area, country_infant, 
                                             country_sdg)

print("Common Countries across all datasets:")
print(common_countries)
print(len(common_countries))
```

```python
all_countries = (country_bmi | country_gdp | country_life | country_male | 
                 country_old | country_young | country_population | 
                 country_urban | country_area | country_infant | 
                 country_sdg)
non_common_countries = all_countries - common_countries

print("Countries not in common across all datasets:")
print(non_common_countries)
```

```python
with open("not_needed.txt", "r",encoding="UTF-8") as file:
    not_needed = {line.strip() for line in file}
bad_countries = [i for i in non_common_countries if i not in not_needed]
bad_countries = sorted(bad_countries, key=lambda x: -len(x))
print(len(bad_countries))
for i in bad_countries[:8]:
    print(i)
```

```python
m1 = pd.merge(young_ratio, old_ratio, on=["Country","Year"], how='inner')
m2 = pd.merge(m1, sdg_regions, left_on=["Country"], right_on=["Country/area"], how='inner')
m2.drop(columns="Country/area",inplace=True)
m2.rename(columns={'2024': 'SDG Region'}, inplace=True)
m3 = pd.merge(m2, population, on=["Country","Year"], how='inner')
m4 = pd.merge(m3, male_ratio, on=["Country","Year"], how='inner')
m5 = pd.merge(m4, urban_population_ratio, on=["Country","Year"], how='inner')
m6 = pd.merge(m5, area, on=["Country","Year"], how='inner')
m7 = pd.merge(m6, gdp_per_capita, on=["Country","Year"], how='inner')
m8 = pd.merge(m7, life_expectancy, on=["Country","Year"], how='inner')
m9 = pd.merge(m8, infant_mortality, left_on=["Country","Year"], right_on=["Country or Area","Year(s)"], how='inner')
m9.drop(columns=["Country or Area","Year(s)"],inplace=True)
m10 = pd.merge(m9, bmi, on=["Country","Year"], how='inner')
m10
```

```python

from itertools import combinations
def check_unique_column_pairs(df):
    unique_pairs = {}
    for col1, col2 in combinations(df.columns, 2):
        is_unique = df.duplicated(subset=[col1, col2]).sum() == 0
        unique_pairs[(col1, col2)] = is_unique
    return unique_pairs

unique_pairs_result = check_unique_column_pairs(m10)

for pair, is_unique in unique_pairs_result.items():
    print(f"Columns {pair} are unique: {is_unique}")
```
