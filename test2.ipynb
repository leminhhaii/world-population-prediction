```python
data = pd.read_csv('latest_country_data.csv')
data.info()
```

```python
data['Population_density(People/sq km)'] = data['Population']/data['Total Area (sq km)']
data.drop('Country', inplace = True, axis=1)
data.drop('Total Area (sq km)', inplace=True, axis=1)
data = pd.get_dummies(data, columns=['SDG Region'], drop_first=True, dtype="float")
numerical_cols = ['Year', 'Young Age Dependency Ratio', 'Retirement Age Dependency Ratio',
       'Population', 'Male Population Ratio', 'Urban Population Ratio',
       'GDP_per_Capita', 'Life Expectancy',
       'Infant mortality per 1000 live births', 'BMI', 'BMI Female',
       'BMI Male', 'Population_density(People/sq km)']
scaler = StandardScaler()
data[numerical_cols] = scaler.fit_transform(data[numerical_cols])

```

```python
data.head()
```

```python
population_linear_model = joblib.load('population_linear.pkl')
```

```python
data1 = pd.read_csv('cleaned_data.csv')
latest_data = data1.loc[data1.groupby('Country')['Year'].idxmax()]
```

```python
country_year_population = latest_data[['Country','Year','Population']]
```

```python
predictions = []
for i in range(187):
    test_value = data.iloc[i].values.reshape(1,-1)
    prediction = population_linear_model.predict(test_value).item()
    predictions.append(prediction)
predictions
```

```python
prediction_file = "prediction_population.csv"
```

```python
country_year_population['population_growth'] = predictions
country_year_population
```

```python
new_rows = []

for country in country_year_population['Country'].unique():
    country_data = country_year_population[country_year_population['Country'] == country]
    
    # Get the last row for each country (last year's data)
    last_year_data = country_data.iloc[-1]
    
    # Calculate the new year and population
    new_year = last_year_data['Year'] + 1
    new_population = last_year_data['Population'] * ((100 + last_year_data['population_growth'])/100)
    
    

    # Create a new row
    new_row = {
        'Country': country,
        'Year': new_year,
        'Population': new_population,
        'population_growth': last_year_data['population_growth']  # Assuming growth rate remains the same
    }
    
    # Append the new row to the list of new rows
    new_rows.append(new_row)

# Convert the new rows list to a DataFrame
new_rows_df = pd.DataFrame(new_rows)

# Append the new rows to the original dataset
country_year_population = pd.concat([country_year_population,new_rows_df], ignore_index=True)

# # Save the updated dataset
country_year_population.to_csv("updated_dataset.csv", index=False)

# # Display the updated dataset
# print(data)
country_year_population
```
